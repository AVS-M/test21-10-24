```python
import requests
from bs4 import BeautifulSoup

# Указываем URL страницы, которую будем парсить
url = 'https://romi.center/ru/learning/article/what-is-Yandex-Metrika-beginners-guide/'

# Отправляем GET-запрос к странице
response = requests.get(url)

# Проверяем, что запрос выполнен успешно
if response.status_code == 200:
    # Инициализируем BeautifulSoup для парсинга HTML-кода
    soup = BeautifulSoup(response.text, 'html.parser')

    # Извлекаем заголовок статьи
    title = soup.find('h1').text
    print(f"Заголовок статьи: {title}")

    # Извлекаем все параграфы текста
    paragraphs = soup.find_all('p')
    for idx, p in enumerate(paragraphs, start=1):
        print(f"Параграф {idx}: {p.text}")

    # Извлекаем все ссылки (теги <a>)
    links = soup.find_all('a')
    for link in links:
        print(f"Ссылка: {link.get('href')} - Текст: {link.text}")

else:
    print(f"Ошибка при выполнении запроса: {response.status_code}")
```
